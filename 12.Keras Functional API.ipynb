```python
#import libraries
import keras
import numpy as np
import pandas as pd
from keras.layers import Input, Dense
from keras.models import Model
from keras import layers, optimizers, losses, metrics
from keras.utils.np_utils import to_categorical


# load dataset
dataframe = pd.read_csv("iris.csv", header=None)
dataset = dataframe.values


# shuffle data to shuffle rocks & mine data
np.random.shuffle(dataset)


# split into input (X) and output (Y) variables
x = dataset[:,0:4].astype(float)
y = dataset[:,4]


#normaliazing data
x -= x.mean(axis=0)
x /= x.std(axis=0)


#encode label in 1 and 0 and 2
a = y.tolist()
b= []
for i in a:
    if i == 'Iris-setosa':
        i = 1
        b.append(i)
    elif i == 'Iris-versicolor':
        i = 2
        b.append(i)
    else:
        i = 0
        b.append(i)
y = np.asarray(b)


#split data in test and train
x_test = x[:30]
x_train = x[30:]
y_test = y[:30]
y_train = y[30:]


# split validation data
x_val = x_train[:30]
partial_x_train = x_train[30:]
y_val = y_train[:30]
partial_y_train = y_train[30:]


#one hot encoding test, train and validation labels
y_test = to_categorical(y_test)
y_val = to_categorical(y_val)
partial_y_train = to_categorical(partial_y_train)


# Functional API 
inputs = Input(shape=(4,))
outputs = layers.Dense(8, activation='relu')(inputs)
predictions = layers.Dense(3, activation='softmax')(outputs)
sgd = optimizers.RMSprop(learning_rate=0.001, rho=0.9)
loss = losses.categorical_crossentropy
metrics = [metrics.categorical_accuracy]
model = Model(inputs=inputs, outputs=predictions)

#compiler

model.compile(sgd,loss,metrics)

#train model
history = model.fit(partial_x_train,partial_y_train,epochs=200,batch_size=5,validation_data=(x_val, y_val))
```

```python
#evaluate on test data
results = model.evaluate(x_test, y_test)
print(results)
```
